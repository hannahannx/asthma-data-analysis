# Multiple Linear Regression 

## Importing Modules

```python
import pandas as pd
from sklearn import linear_model
```

# Load data set

```python
df = pd.read_csv('../data/cleaned_data/cleaned_datasetIMPORTED.csv.csv')
```

## Independent and Dependent Varibles

```python
X = df[["Emergency Medical Admissions","FY1/2 - Total Repository Staff Posts"]]
y = df[["FY1/2 -Filled Respiratory Staff Posts"]]
```

## Relationship


```python
regr = linear_model.LinearRegression()
regr.fit(X.values, y)
```

## Predict the amount of Respitay staff needed

```python
predictions = regr.predict(1000,)
print("Predicted Staff needed ", predictions) # round the number up because you cant hav e afloat of a staff
```
