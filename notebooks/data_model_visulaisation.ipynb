# Data Modelling and Visualisation

# Multiple Linear Regression 

## Importing Modules

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

## Load data set

```python
df = pd.read_csv('../data/cleaned_data/cleaned_datasetIMPORTED.csv')
```

## Independent and Dependent Varibles

```python
# Independent
X = df[["Paediatric Asthma Admissions Per 1000 Admissions","FY1/2 - Total Repository Staff Posts"]]
# Dependent
y = df[["FY1/2 -Filled Respiratory Staff Posts"]]
```

Training and Test splits

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Relationship


```python
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)
```

## Predict the amount of Respiratory staff needed

```python
y_predictions = regr.predict(X_test)
print("Predicted Filled Staff Posts ", y_predictions) # round the number up because you cant hav e afloat of a staff
```

```python
print("Coefficients of Asthma admissions per 1000 and total respitory staff posts againsed the total number of filled post :", regr.coef_)
```

```python
plt.scatter(y_test, y_predictions)
plt.plot(y_test, y_predictions)
plt.show()


plt.scatter(y_predictions, y_predictions)
plt.plot(y_test, y_predictions)
plt.show()
```
